```python
import numpy as np

a = 2
b = 3

x_line = np.linspace(0, 10, 100)
y_pred = a * np.array(x_line) + b

plt.scatter(x, y)
plt.plot(x_line, y_pred)
plt.show()
```

```python
import numpy as np

order = 4  # (order)차 함수를 사용
mymodel = np.poly1d(np.polyfit(x, y, order))  # 회귀 문제를 알아서 풀어줌!
print(mymodel)

# 우리가 찾은 모델을 그려보자
myline = np.linspace(0, 10, 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```
